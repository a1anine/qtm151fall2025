1. Use an if/elif/else block to test whether a string belongs to one of two lists. Depending on the outcome (belongs to list A, belongs to list B, doesn't belong to either list) print an appropriate message to the screen.

```python
list_A = ["apple", "banana", "cherry"]
list_B = ["dog", "cat", "rabbit"]

# Define the two lists
list_A = ["apple", "banana", "cherry"]
list_B = ["dog", "cat", "rabbit"]

# Ask user for input
word = input("Enter a word: ")

# Check which list the word belongs to
if word in list_A:
    print("The word belongs to list A.")
elif word in list_B:
    print("The word belongs to list B.")
else:
    print("The word does not belong to either list.")
```

2. You are given two lists of numbers, x_vals and y_vals, and must create a Numpy array where the components are related to those in the lists (for example, you might be asked to create a Numpy array with entries 
z_k = (x_k)^2-5sin(y_k)). Then generate a histogram of the values in the Numpy array, with labeled axes and a title. You will also be asked to change the color of the bars in the histogram.

```python
x_vals = [1, 2, 3, 4, 5]
y_vals = [2, 4, 6, 8, 10]

import numpy as np
import matplotlib.pyplot as plt

# Define two lists of numbers
x_vals = [1, 2, 3, 4, 5]
y_vals = [2, 4, 6, 8, 10]

# Create a Numpy array where each element is the sum of corresponding x and y values
arr = np.array(x_vals) + np.array(y_vals)

# Generate a histogram of the array
plt.hist(arr, color="green", edgecolor="black")

# Label axes and add a title
plt.xlabel("Sum of x and y values")
plt.ylabel("Frequency")
plt.title("Histogram of x + y")

# Display the histogram
plt.show()
```

3. Generate random samples from two specified distributions and plot their histograms, side by side in a single figure. Each sub-figure will have its own labeled axes and title.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples
normal_samples = np.random.normal(0, 1, 1000)
uniform_samples = np.random.uniform(0, 1, 1000)

# Create figure with two subplots
plt.figure(figsize=(10, 4))

# Plot normal distribution histogram
plt.subplot(1, 2, 1)
plt.hist(normal_samples, bins=20, color="skyblue", edgecolor="black")
plt.title("Normal Distribution (μ=0, σ=1)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plot uniform distribution histogram
plt.subplot(1, 2, 2)
plt.hist(uniform_samples, bins=20, color="lightcoral", edgecolor="black")
plt.title("Uniform Distribution (0,1)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Adjust layout and display
plt.tight_layout()
plt.show()
```

6. Create a list of .csv file names and an additional empty list. Then in a for loop, read in each file to make a DataFrame, compute the mean of one of the variables/features/columns, and store it as an element of the originally empty list. Then print the list. 

```python
# List of CSV file names
list_datasets = ["dataset1", "dataset2", "dataset3"]

import pandas as pd
# Create an empty list to store mean data
list_data = []

# Loop through each dataset in the list
for file in list_datasets:
    # Read the dataset
    data = pd.read_csv(file)
    
    # Compute mean of column
    mean_data = data[""].mean()
    
    # Append the result to list_data
    list_data.append(mean_data)

# Print the list of computed means
print(list_data)
```
